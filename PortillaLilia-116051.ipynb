```python
#2. Leer dos números enteros y determinar si la magnitud de la diferencia entre los dos es un número primo. 



import math

print("LA DIFERENCIA ENTRE DOS NUMEROS ES O NO PRIMO")
#pide los numeros
num1=int(input("Ingrese el primer numero: "))
num2=int(input("Ingrese el segundo numero: "))
#calcula la diferencia
diferencia=abs(num1-num2)
print("La diferencia entre ",num1," y ",num2," es ",diferencia)

if diferencia<2:#si la diferencia es  menor a 2 no es primo
    print(diferencia," no es primo")
elif diferencia==2 or diferencia==3:#si la diferencia es 2 o 3 es primo
    print(diferencia," es primo")
else:# si no es ni menor a 2 ni es 2 o 3
    i=2#inicia variable de recorrido
    while i<=math.sqrt(diferencia):#ciclo que termina cuando se lle a la raiz 2 de la diferencia
        if diferencia%i==0:#diferencia par(se descartan)
            print(diferencia," no es primo")#no es primo
            break#rompe el ciclo
        if i==2:#si va en la segunda posicion
            i+=1#se itera en 1
        else:#si el indice es impar
            i+=2#se itera de a 2 para descartar pares		
        if i>math.sqrt(diferencia):#si el indice es mayor a la diferencia
            print(diferencia," es primo")
```

```python
#5. Escribir un programa que reciba una cadena de texto y reporte: 
#a) cuántas letras vocales en mayúscula se entraron;
#b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
#c) cuántos dígitos se entraron,  
# d) cuántos espacios se entraron;  e) cuántas palabras reservadas se entraron.  
 
#Utilizar para ello los métodos de cadenas especializados como .isupper(), .isdigit(), etc. y la palabra reservada in.  
 
#libreria 
import keyword

#Titulo
print("VOCALES MAYUSCULAS, TILDES, DIGITOS, ESPACIOS Y PALABRAS RESERVADAS")
print("")

cadena=input("Ingrese la cadena de texto: ")#se pide la cadena
palabras=cadena.split()#se sacan las palabras de la cadena
lista_reservadas=keyword.kwlist#guarda todas las palabras reservadas en una lista

#inicia variables contadoras para sacar los datos
vocales=0
tildes=0
digitos=0
espacios=0
reservadas=0
for i in cadena:#recorre la cadena
    if i in "AEIOU":#pregunta si hay vocales mayusculas
        vocales=vocales+1#se suma una vocal 
    if i in "´áéíóúýÁÉÍÓÚÝ":#si tiene tilde
        tildes=tildes+1
    if i.isdigit():#si es digito
        digitos=digitos+1
    if i.isspace():#si es un espacio
        espacios=espacios+1
for i in palabras:#se recorre las palabras
    if i in lista_reservadas:#recorre lalista de palabras reservadas
        reservadas=reservadas+1

#muestra resultados
print("En la cadena ingresada: ")
print("Hay ",vocales," vocales mayusculas")
print("Hay ",tildes, " tldes")
print("Hay ",digitos," digitos")
print("Hay ",espacios," espacios")
print("Hay ",reservadas," palabras reservadas")
```

```python
#10.Leer una matriz e imprimir el vector que aparece cuando la matriz se desenreda en forma de espiral que gira en el sentido horario a partir de la esquina inferior derecha.

#Tmaño de la matriz(mxn), es igresado por el usuario 
print("Tamaño de la matriz")
m=int(input("De cuantas filas es la matriz: "))
n=int(input("De cuantas columnas es la matriz: "))

#Iniciando la matriz con elementos nulos
matriz = [[None] * n for i in range(m)]
#Agregar elementos a la matriz, por parte de el usuario
for i in range(m):
    for j in range(n):
        matriz[i][j]=int(input(f"Ingrese el elemento en la fila{i+1}, columna{j+1}: "))

#Mostrar matriz ingresada
print("la mtriz ingresada es: ")
for i in range(m):
    print("")
    for j in range(m):
        print(matriz[i][j],end="  ")
print("")

#Variable donde se guarda el vector de la espiral
vec=[]
#variables con los distintos puntos geograficos en la matriz
filai=0#primer fila
filaf=m-1#ultima fila
columnai=0#primer columna
columnaf=n-1#ultima columna

#Se ira recorriendo la matriz desde el marco mas exterior a el mas interior
#esto sehara acercando la fila y columna inicial a la fila y columna final
#Ciclo que se finalizara cuando no haya mas marcos que recorrer es decir, 
#cuando la fila y columna inicial sean mayores que la fila y columna final
while (filai<=filaf) and (columnai<=columnaf):

    for j in range(columnaf,columnai-1,-1):#ultima fila, y recorre las columnas hacia atras
        vec.append(matriz[filaf][j])#agregar elementos al vector
    for i in range(filaf-1,filai-1,-1):#primer columna, y recorre las filas desde la penultima hasta la primera
        vec.append(matriz[i][columnai])#agregar elementos al vector
    for j in range(columnai+1,columnaf+1,1):#primer fila, y recorre las columnas desde la segunda hasta la ultima (incluida)
        vec.append(matriz[filai][j])#agregar elementos al vector
    for i in range(filai+1,filaf,1):#ultima columna, y recorre las filas desde la segunda hasta la penultima
        vec.append(matriz[i][columnaf])#agregar elementos al vector
    #generamos un marco mas interno	
    filai+=1#reducimos desde arriba
    filaf-=1#reducimos desde abajo
    columnai+=1#reducimos desde izquierda 
    columnaf-=1#reducimos desde derecha


#Mostramos el vector que aparece cuando la matriz se desenreda en forma de espiral que gira en el sentido horario a partir de la esquina inferior derecha.
print("El vector solicitado es el ", vec)
```

```python
#12. Diseñar una función que pida la anchura de un triángulo y lo dibuje con caracteres producto (*). 

#definimos la funcion
def dib_triangulo():
    anchura = int(input("Anchura del triángulo: "))#pedimos la anchura
    for i in range(1, anchura + 1):#ciclo externo que maneja las filas del dibujo empezando con un elemento, hasta la mitad donde alcanza el maximo de elementos(anchura)
        for j in range(i):#asigna cuantos elementos van en cad fila
            print("* ", end="")#imprime el caracter(es) de la fila
        print()#salto cuando pasa a la siguiente fila

    for i in range(1, anchura):#parte inferior del triangulo
        for j in range(anchura - i):
            print("* ", end="")
        print()


dib_triangulo()#llamamos la funcion
```

```python
#14. Diseñar una función que proponga sumas de números positivos (dos números entre 1 y 100) al usuario y compruebe la respuesta.
#El programa continuará hasta que se acierten cinco sumas. (Tip: para generar los números enteros aleatorios, existe la función random.randint(a,b) de la librería random). 



# Funcion que propone sumas de numeros positivos al usuario y comprueba la respuesta y se ejecuta hasta que el usuario acierte 5 sumas.

import random # Importamos la libreria random que nos permite hacer uso de funciones para generar numeros aleatorios.


print("Escriba el resultado de las siguientes operaciones:") # Imprimimos un mensaje para el usuario, indicandole lo que debe hacer.

def sum_positivos():
    contador=0 # Inicializamos en 0 la variable contador que se encargara de contar las respuestas correctas.
    while contador<5: #Inicializamos un ciclo while que se ejecutara hasta que el contador sea igual a 5. Proponiendo nuevas sumas en cada ejecucion.
        a = random.randint(1, 100) # Guardamos en la variable a un numero aleatorio entre 1 y 100 incluidos estos. El numero se genera con la funcion random.randint()
        b = random.randint(1, 100) # Guardamos en la variable b un numero aleatorio entre 1 y 100 incluidos estos. El numero se genera con la funcion random.randint()
        suma= a+b #Guardamos en la variable suma, la suma de de los numeros aleatorios a y b. 
        respuesta = int(input(f"{a} + {b} = ")) #Guardamos en la variable respuesta, la respuesta dada por el usuario. Esto lo hacemos con la funcion input que recibe datos del usuario, en este caso un numero entero(int).
        if respuesta==suma: # Iniciamos la sentencia condicional if, y preguntamos si la respuesta dada por el usuario(respuesta) es igual la suma de los numeros aleatorios hecha en el programa(variable "suma")
            contador = contador+1 # Si la respuesta del usuario es correcta contamos una suma correcta mas. Lo hacemos sumandole uno al contador.
            print("¡Respuesta correcta!") # Imprimimos un mensaje diciendole al usuario que la respuesta es correcta.
        else: # En caso de que la respuesta dada por el usuario sea incorrecta, tomamos este camino.
            print("¡Respuesta incorrecta!") # Imprimimos un mensaje diciendole al usuario que la respuesta es incorrecta.
        # Si el contador es menor que 5 se repite el proceso, propomiendo una nueva suma. En caso contrario el ciclo while se rompe(termina).

    print("Programa terminado")	# Imprimimos un mensaje diciendole al usuario que el programa ha terminado.


sum_positivos()#llamaos la funcion

```

```python
#15. Diseñar una función que permita jugar a una versión simplificada del juego Master Mind.
#El juego consistirá en adivinar una cadena de números distintos.
#La función debe pedir la longitud de la cadena (de 2 a 9 cifras) y generar aleatoriamente un número con esta cantidad de cifras;
#después el programa debe ir pidiendo que intente adivinar la cadena de números. En cada intento,
#el programa informará de cuántos números han sido acertados (el programa considerará que se ha acertado un número si coincide el valor y la posición). 


#libreria par generar numeros aleatorios
import random

def master_mind():#definimos funcion(sin parametros)

    longitud=1#Inicializamos la longitud
    while longitud<2 or longitud>9:#exigimos que la longitud este en el rango de 2 a 9
        longitud=int(input("Digame la longitud de la cadena (entre 2 y 9): "))#pide la longitud

    num=random.randint(10**longitud/10,10**longitud-1)#se genera el numero aleatorio con numero de cifras igual a la longitud
    cad=str(num)#convierte el numero en cadena
    adivinar=None#iniciamos la variable donde se guardan las respuestas del usuario, inicia sin valor
    contador=0#contador de aciertos en 0

    while contador<longitud:#ciclo que se cierra cuando el usuario adivine los digitos
        adivinar=input("Intente adivinar la cadena: ")#pide la respuesta
        for i in range(0,longitud):#recorre la cadena de el digito aleatorio y la cadena de respuesta, buscando aciertos 
            if adivinar[i]==cad[i]:#si tanto digito como ubicacion son iguales
                contador+=1#se suma el acierto
        if contador==longitud:#si adivino la cadena de numeros
            print("Con ",adivinar," ha adivinado ",contador," valores.")#muestra el numero de aciertos
            print("Felicidades")#lo felicita por adivinar la cadena de numeros
        else:#si no logro adivinar la cadena de numeros
            print("Con ",adivinar," ha adivinado ",contador," valores.",end=" ")#le muestra cuantos aciertos tuvo
            contador=0#se inicia el contador en cero nuevamente


master_mind()#se llama la funcion
```
